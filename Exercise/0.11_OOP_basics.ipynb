````python
#0.)


class converting:
    def __init__(self, value) -> None:
        if not isinstance(value, (int, float)):
            raise TypeError("Only integer or float values are allowed")  
        self.value = value

    @property
    def value(self):
        return self._value

    @value.setter
    def value(self, new_value):
        if not isinstance(new_value, (int, float)):
            raise TypeError("Only integer or float values are allowed")
        self._value = new_value

    def inch_to_cm(self):
        return self.value*2.54       
    
    def foot_to_meter(self):
        return self.value*0.3048
    
    def pound_to_kg(self):
        return self.value*0.453592

    def __repr__(self):
        return f"Converting(value={self.value})"

#Example usage
c1 = converting(20)
print(c1.foot_to_meter())  # This will print the conversion from feet to meters

# Testing property setter
c1.value = 5
print(c1.foot_to_meter())  # This should now print the conversion for 5 feet to meters

````

````python
#1.)

class Person:
    def __init__(self, name: str, age: int, email: str) -> None:
        if not isinstance(name, (str)):
            raise ValueError("Only string is allowed")
        if not isinstance(age, (int)):
            raise ValueError("Only int is allowed")
        if not "@" in email:
            raise ValueError("must contain @")
        self._name = name
        self._age = age
        self._email = email

    @property
    def name(self):
        return self._name ##### ______________name...

    @property
    def age(self):
        return self._age ######______________age

    @property
    def email(self):
        return self._email ######___________email
    
    def say_hello(self):
        return f"Hi, my name is {self._name}, I am {self._age} years old, my email address is {self._email}"

    def __repr__(self):
        return f"Person(name={self._name}, age={self._age}, email={self._email})"

# Example usage
p1 = Person("Börje", 26, "borje@bajs.se")
print(p1.say_hello())

````
