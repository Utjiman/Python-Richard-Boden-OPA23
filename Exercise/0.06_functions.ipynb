````python
# 0.)
def Triangle(base, height):
    area = base * height / 2
    return area 
Triangle(18, 6)
````

````python
# 1.a)
import math
def euclidean(a, b):
    x1, y1 = a
    x2, y2 = b
    distance = math.sqrt((x2 -x1)**2 + (y2-y1)**2)
    return distance

a = (4, 4)
b = (2, 6)

print(f"{a} {b} -> {euclidean(a ,b)}")
````

````python
#1 .b)
#  b)   Let the user input two points. Call the function using the users input points. (*)

print("Enter the coordinates for the first point:")
x1 = float(input("Input x coordinate for the first point: "))
y1 = float(input("Input y coordinate for the first point: "))

print("Enter the coordinates for the second point:")
x2 = float(input("Input x coordinate for the second point: "))
y2 = float(input("Input y coordinate for the second point: "))

point1 = (x1, y1)
point2 = (x2, y2)

distance = euclidean(point1, point2)
print(f"distance between points {distance}")
````

````python
#1 .c)

points = [(10, 3), (-1, -9), (10, -10), (4, -2), (9, -10)]
origin = (0, 0)

distances = [euclidean(origin, point) for point in points]
distances
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
f = lambda x: x**2 - 3

# Generate x values
x = np.linspace(-10, 10, 400)

# Generate y values for each function
y_f = f(x)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(x, y_f, label='f(x) = x^2 - 3')

# Labeling the plot
plt.title('Graph of f(x) = x^2 - 3')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
````
